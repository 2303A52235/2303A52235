##part-02:lab task  is tovisualize the datasets or data using matplotlib

```python
people =['arjun','rajesh','rahul','sravan']
age=[20,21,19,25]
weight=[65,72,60,80]
height=[5.10,6.1,5.5,5.11]


```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("relation between weight and height of patients")
plt.ylabel("height in feets")
plt.xlabel("weight in Kilograms")
plt.show()
```

```python
plt.scatter(weight,age)
plt.title("relation between age and weight of patients")
plt.ylabel("age in years")
plt.xlabel("weight in Kilograms")
plt.show()
```

## Generate Bar Plot


```python
plt.figure(figsize=(12,4))
plt.title("people weights in kgs", fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.5, color='gold',edgecolor='k',alpha=1)
plt.xlabel('people',fontsize=10)
plt.ylabel('weight(in kgs)',fontsize=10)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='blue',edgecolor='k',alpha=1,bins=5)
plt.title("histrogram of people weights in kgs", fontsize=16,fontstyle='italic')
plt.xlabel('weight(in kgs)',fontsize=10)
plt.ylabel('people',fontsize=10)
plt.show()
```
